#### Статистика

```python
a = np.array([1, 2, 3, 4, 5])
a
```

**Среднее арифметическое**

```python
mean = a.sum() / len(a)
mean
```

**Дисперсия и стандартное отклонение**

*Дисперсия* - это показатель вариации, который представляет собой средний квадрат отклонений от математического ожидания (среднее арифметическое). Она выражает разброс чисел относительно среднего значения.

*Стандартное отклонение* (среднеквадратичное отклонение) - это величина, характеризующая разброс данных относительно среднего значения.

```python
a - mean
```

$$var = \frac{1}{n}\sum(x - x_m)^2$$

```python
variance = np.mean((a - mean) ** 2)
std = np.sqrt(variance)
variance, std
```

Подсчет с помощью numpy

```python
mean_np = np.mean(a)
mean_np
```

```python
std_np = np.std(a)
std_np
```

```python
np.var(a)
```

**Min, Max**

```python
np.min(a), np.max(a)
```

**Медиана**

*Медиана* - значение, которая делит его на две равные части или же проще говоря, это середина отсортированной выборки.

```python
a = [3, 4, 5, 6, 1, 2, 2, 6]
a = np.sort(a)
a
```

*Квартили* - это значения, которые делят выборку на 4 равные (по количеству объектов) части.
Откуда следует, что вторая квартиль - это медиана.

```python
np.quantile(a, 0.5)
```

Мода - это объект с максимальной частотой или проще говоря, самый популярный (частовстречаемый) объект в выборке.

```python
df = pd.read_csv('advertising.csv')

df['User_id'] = pd.Series(range(1000, 2000))

df.head()
```

```python
cols_1 = ['User_id', 'Daily Time Spent on Site', 'Ad Topic Line', 'Timestamp', 'Clicked on Ad']
df_site = df[cols_1]

print('Размер датасета df_site:', df_site.shape)
df_site.head()
```

```python
cols_2 = ['User_id', 'Male', 'Age', 'Country', 'City', 'Daily Internet Usage']
df_personal = df[cols_2]

print('Размер датасета df_personal:', df_personal.shape)
df_personal.head()
```

```python
cols_3 = ['Country', 'City', 'Area Income']

df_macro = df[cols_3][:-50]

print('Размер датасета df_macro:', df_macro.shape)
df_macro.head()
```

### merge

<img src='https://absentdata.com/wp-content/uploads/2019/07/pd.merge-1.png'>

<img src='https://www.thecrazyprogrammer.com/wp-content/uploads/2019/05/Joins-in-SQL-Inner-Outer-Left-and-Right-Join.jpg?ezimgfmt=rs:372x290/rscb1/ngcb1/notWebP'>

```python
# Предположим, что мы получили эти три датасета из разных источников и теперь нам нужно их объединить

df_2 = pd.merge(df_site, df_personal, on='User_id', how='inner')

print('Размер датасета df_2:', df_2.shape)
df_2.head()
```

```python
df_macro.shape
```

```python
df_3 = df_2.merge(df_macro, left_on=['Country', 'City'], right_on=['Country', 'City'], how='left')

print('Размер датасета df_3:', df_3.shape)
df_3.head()
```

```python
print('Частично пустых строк')
df_3.isna().sum()
```

```python
df_3[df_3['Area Income'].isna()].head()
```

```python
df_macro[df_macro['Country'] == 'Colombia']
```

```python
df_4 = df_2.merge(df_macro, on=['Country', 'City'], how='inner')

print('Размер датасета df_4:', df_4.shape)
df_4.head()
```

```python
df_5 = df_2.merge(df_macro, on=['Country', 'City'], how='right')

print('Размер датасета df_5:', df_5.shape)
df_5.head()
```

### pivot_table

```python
df.pivot_table(
    index='Male',
    columns='Clicked on Ad',
    values='Daily Internet Usage',
    aggfunc='mean'
)
```

```python
pd.pivot_table(data=df,
               values='Daily Internet Usage',
               index='Male',
               columns='Clicked on Ad',
               aggfunc='mean')
```

```python
pd.pivot_table(data=df,
               values='Daily Internet Usage',
               index=['Male', 'Clicked on Ad'],
               columns='Country',
               aggfunc='mean')
```
